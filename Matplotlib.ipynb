### Bar_plot

```python
course = ['Python','Java','.NET c#','GO']
revenue=[900,1660,891,247]
```

```python
len(course)
```

```python
a = np.arange(len(course))
a
```

```python
plt.bar(a,revenue,label='Revenue_data',color = 'g')
plt.title('Bar_plot')
plt.xticks(a,course)
plt.ylabel('Revenue')
plt.legend(loc=1 , fontsize='large')
plt.show()
```

```python
plt.barh(a,revenue,color = 'r',label='Revenue')
plt.title('Bar_plot')
plt.yticks(a,course)
plt.xlabel('Revenue')
plt.legend(loc=1)
plt.show()
```

## sub_plots

```python
a=[2014,2015,2016,2017]
b=[4000,4500,5300,4600]
c=[2800,3000,2800,3400]

plt.plot(a , b ,color = 'r' , marker = '+',label = 'A_B')

plt.plot(a , c ,color = 'b',marker = '*',label = 'A_c')
plt.legend(loc = 0)
plt.show()
```

```python
a=[2014,2015,2016,2017]
b=[4000,4500,5300,4600]
c=[2800,3000,2800,3400]

plt.subplot(2,1,2) # 1sr= row , 2nd = columns , 3rd = location we want
plt.plot(a , b ,color = 'r' , marker = '+',label = 'A_B')
plt.legend(loc = 0)
plt.subplot(2,1,1)
plt.plot(a , c ,color = 'b',marker = '*',label = 'A_c')
plt.legend(loc = 0)
plt.show()
```

```python
plt.subplot(1,1,1)
plt.bar(a,b,color='red')
```

## Pie_chart

```python
x = [40,30,20,10,1]
y = ['java','python','C#','PHP','c++']
plt.pie(x,labels=y) #labels means showing 
```

```python
plt.figure(figsize = (10 , 10))
x = [40,30,20,10,1]
y = ['java','python','C#','PHP','c++']
plt.pie(x,labels=y )
```

```python
plt.figure(figsize = (7 , 7))
x = [40,30,20,10,1]
y = ['java','python','C#','PHP','c++']
plt.pie(x,labels=y ,shadow = True ,autopct='%1.1f%%',radius = 1.25)
```
